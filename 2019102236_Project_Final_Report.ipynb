```python
def franchises():
    df = pandas.read_csv('List of highest grossing video game franchises.csv')
    df['Rank'] = np.arange(1,71)
    df = pandas.DataFrame(df, columns=['Rank','Franchise','Year of inception','Lifetime revenue','Franchise owner(s)','Platform(s) of origin'])
    df = df[df['Franchise owner(s)']=='Activision']
    df = df[df['Franchise']=='Call of Duty']
        
    return df


def best_selling_games():
    df = pandas.read_csv('List of best-selling video games of all time.csv')
    df['Rank'] = np.arange(1,51)
    df = pandas.DataFrame(df, columns=['Rank','Title','Sales','Initial release date','Developer(s)','Publisher(s)'])
    df = df[df['Publisher(s)']=='Activision']

    return df

def best_selling_Xbox_one_games():
    df = pandas.read_csv('List of best selling Xbox one games.csv')
    df['Rank'] = np.arange(1,18)
    df = pandas.DataFrame(df, columns=['Rank','Title','Total copies sold', 'Sales BreakDown', 'Release date','Genre', 'Developer', 'Publisher'])
    df = df.drop(['Sales BreakDown'],axis=1)
    df = df[df['Publisher']=='Activision']

    return df

def best_selling_Xbox_360_games():
    df = pandas.read_csv('List of best selling Xbox 360 games.csv')
    df['Rank'] = np.arange(1,64)
    df = pandas.DataFrame(df, columns=['Rank','Title','Total copies sold', 'Sales BreakDown', 'Release date','Genre', 'Developer', 'Publisher'])
    df = df.drop(['Sales BreakDown'],axis=1)
    df = df.drop(['Release date'],axis=1)
    df = df[df['Publisher']=='Activision']
    df = df[df['Genre']== 'First-person shooter']

    return df


def best_selling_PS4():
    df = pandas.read_csv('List of best selling PS4.csv')
    df['Rank']=np.arange(1,29)
    df = pandas.DataFrame(df, columns=['Rank','Title','Total copies sold', 'Genre', 'Developer', 'Publisher'])
    df = df[df['Publisher']=='Activision']
    df = df[df['Genre']=='First-person shooter']

    return df



def best_selling_PS3():
    df = pandas.read_csv('List of best selling PS3.csv')
    databox = []
    databox1 = {'Rank':[], 'Title' : [],'Copies sold' : []}
    files=open("List of best selling PS3.csv","r")
    for line in files:
        databox.append(line.strip('\n').split(','))
    files.close()
    n=-1
    for line2 in databox:
        for i in line2:
            if i.find("million") >=0:
                continue
            else:
                n += 1
            if i.find("Call of Duty") >= 0:
                databox1['Rank'].append(n-1)
                databox1['Title'].append(line2[0])
                databox1['Copies sold'].append(line2[1])
                break
    df = pandas.DataFrame(databox1)

    return df


def most_played_video_game():
    df = pandas.read_csv('List of most played video game.csv')
    df['Rank'] = np.arange(1,85)
    df = pandas.DataFrame(df, columns=['Rank','title','Number', 'As of', 'Business model', 'Release date','Publisher'])
    df = df.drop(['Business model'],axis=1)
    df = df[df['Publisher']=='Activision']

    return df
```
